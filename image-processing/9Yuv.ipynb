```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
image_path = os.path.join("../images", "salah.jpg")
image = cv2.imread(image_path)

# function to display the image
def show_image(title,img,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

show_image("Original image (BGR)",image)
```

```python
import cv2

# Convert image to YUV color space
image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

# Function to display images (assuming show_image is predefined)
def show_image(title, img):
    import matplotlib.pyplot as plt
    plt.figure(figsize=(6, 6))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert to RGB for proper visualization
    plt.title(title)
    plt.axis("off")
    plt.show()

# Show YUV image converted back to RGB
show_image("Y'UV (Converted to RGB)", cv2.cvtColor(image_yuv, cv2.COLOR_YUV2BGR))

# Function to display individual color channels
def show_channels(title, img, channel_names, color_space="YUV"):
    channels = cv2.split(img)
    for i, channel in enumerate(channels):
        cmap = "gray" if color_space != "RGB" else None  # HSV and YUV should be shown in grayscale
        show_image(f"{title} - {channel_names[i]}", channel)

# Show YUV channels
show_channels("Y'UV Channels", image_yuv, ["Luma (Y')", "Chrominance U", "Chrominance V"], color_space="YUV")
```
