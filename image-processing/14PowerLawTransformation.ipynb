```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np

# Define the image path
image_path = os.path.join("../images", "salah.jpg")

# Load the image in BGR format (default in OpenCV)
image_bgr = cv2.imread(image_path)

# Convert BGR to RGB for correct color visualization in Matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Load the grayscale image
gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Function to display the image
def show_image(title, img, cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

# Display the image in RGB format
show_image("Original Image (RGB)", image_rgb)
```

```python
gamma = 2.0  # Usually between 0.1 and 2.5 for better results

# Normalize and apply power-law transformation
gray_image = gray_image.astype(np.float32) / 255.0  # Normalize to range [0,1]
power_law_image = np.power(gray_image, gamma) * 255  # Apply gamma correction

# Convert back to uint8
power_law_image = np.clip(power_law_image, 0, 255).astype(np.uint8)
```

```python
show_image("Power law image", power_law_image)
```
