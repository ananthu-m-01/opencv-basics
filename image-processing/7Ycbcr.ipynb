```python
import cv2
import matplotlib.pyplot as plt
import os

image_path = os.path.join("../images", "salah.jpg")
image = cv2.imread(image_path)

# function to display the image
def show_image(title,img,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

show_image("Original image (BGR)",image)
```

```python
def show_channels(title, img, channel_names, color_space="RGB"):
    channels = cv2.split(img)
    
    for i, channel in enumerate(channels):
        cmap = "gray" if color_space != "HSV" else None  # HSV can be displayed as-is
        show_image(f"{title} - {channel_names[i]}", channel, cmap=cmap)
```

```python
image_ycbcr = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

# Show Y'CbCr image converted back to RGB
show_image("Y'CbCr (Converted to RGB)", cv2.cvtColor(image_ycbcr, cv2.COLOR_YCrCb2RGB))

# Show individual Y'CbCr channels
show_channels(
    "Y'CbCr Channels",
    image_ycbcr,
    ["Luma (Y')", "Chroma Blue (Cb)", "Chroma Red (Cr)"]
)
```
