```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
image_path = os.path.join("..","../images","salah.jpg")
image = cv2.imread(image_path)
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
hsv_image = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
```

```python
def show_image(img,title,cmap=None):
    plt.figure(figsize=(3,3))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
show_image(rgb_image,"RGB")
```

```python
def show_channel(img,title,channel_names,color_space="RGB"):
    channels = cv2.split(img)
    for i,channel in enumerate(channels):
        cmap = "gray" if color_space != "HSV" else None
        show_image(channel,f"{title} {channel_names[i]}",cmap=cmap)
```

```python
channel_names = ["Hue","Saturation","Values"]
show_channel(hsv_image,"HSV Channel",channel_names,"HSV")
```
