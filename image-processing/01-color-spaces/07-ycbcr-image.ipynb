```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
image_path = os.path.join("..","../images","salah.jpg")
image = cv2.imread(image_path)
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
ycr_image = cv2.cvtColor(image,cv2.COLOR_BGR2YCrCb)
```

```python
def show_image(img,title,cmap=None):
    plt.figure(figsize=(3,3))
    plt.axis("off")
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.show()
```

```python
def show_channels(img,title,channel_names,color_space="RGB"):
    channels = cv2.split(img)
    for i,channel in enumerate(channels):
        cmap = "gray" if color_space != "YCrCb" else None
        show_image(channel,f"{title} {channel_names[i]}",cmap=cmap)
```

```python
show_image(rgb_image,"RGB")
```

```python
channel_names = ["Luma","Chrominance Blue","Chrominance Red"]
show_channels(ycr_image,"YCrCb",channel_names,"YCrCb")
```
