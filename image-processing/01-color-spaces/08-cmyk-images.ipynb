```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
image_path = os.path.join("..","../images","salah.jpg")
image = cv2.imread(image_path)
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cmyk_image = 1 - rgb_image.astype(float) / 255.0
```

```python
def show_image(title,img,cmap=None):
    plt.figure(figsize=(3,3))
    plt.axis("off")
    plt.title(title)
    plt.imshow(img,cmap=cmap)
    plt.show()
```

```python
K = np.min(cmyk_image,axis=2)
C = (cmyk_image[:,:,0]-K) / (1-K+1e-6)
M = (cmyk_image[:,:,1]-K) / (1-K+1e-6)
Y = (cmyk_image[:,:,2]-K) / (1-K+1e-6)
CMYK = np.dstack((C,M,Y,K))
```

```python
show_image("CMYK - Cyan", C, cmap="Blues")
show_image("CMYK - Magenta", M, cmap="Reds")
show_image("CMYK - Yellow", Y, cmap="Oranges")
show_image("CMYK - Black (K)", K, cmap="gray")
```
