```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
image_path = os.path.join("..","../images","messi.webp")
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
def show_image(img,title,cmap=None):
    plt.figure(figsize=(3,3))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
show_image(image_rgb,"RGB_image")
```

```python
red,green,blue = cv2.split(image)
```

```python
merged_image = cv2.merge((green,red,blue))
show_image(merged_image,"Merged GRB")
```
