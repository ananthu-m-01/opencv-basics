```python
import matplotlib.pyplot as plt
import cv2
import os
```

```python
image_path = os.path.join("..","../images","messi.webp")
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
def show_image(img,title,cmap=None):
    plt.figure(figsize=(3,3))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
show_image(image_rgb,"Original RGB Image")
```

```python
red,green,blue = cv2.split(image)
```

```python
show_image(red,"RED Channel",cmap="Reds")
```

```python
show_image(green,"GREEN Channel",cmap="Greens")
```

```python
show_image(blue,"BLUE Channel",cmap="Blues")
```
