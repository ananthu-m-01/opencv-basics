```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
image_path = os.path.join("..","../images","messi.webp")
image = cv2.imread(image_path)
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
def show_image(img,title,cmap=None):
    plt.figure(figsize=(3,3))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
_,black_and_white = cv2.threshold(image_gray,130,255,cv2.THRESH_BINARY)
show_image(black_and_white,"Black & White",cmap="gray")
```
