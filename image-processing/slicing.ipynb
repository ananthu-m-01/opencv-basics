```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

image_path = os.path.join("../images", "salah.jpg")
gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
bit_planes = []
for i in range(8):
    bit_plane = (gray_image >> i) & 1 
    bit_planes.append(bit_plane * 255)
```

```python
def show_image(img,title,cmap=None):
    plt.figure(figsize=(3,3))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
show_image(bit_planes[0],"Bitplane 1",cmap="gray")
show_image(bit_planes[7],"Bitplane 8",cmap="gray")
```

```python
gamma_value=2.2
gamma_correction=np.array(gray_image**(1/gamma_value),dtype='uint8')
show_image(gamma_correction,"gc",cmap="gray")
```
