```python
import cv2
import matplotlib.pyplot as plt
import os

image_path = os.path.join("../images", "messi.webp")
image = cv2.imread(image_path)

# function to display the image
def show_image(title,img,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

show_image("Original image (BGR)",image)
```

```python
# converting the image into RGB form
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
show_image("Original image(RGB)",image_rgb)
```

```python
print("pixel value in BGR")
print(image[10,10])
```

```python
print("pixel value in RGB")
print(image_rgb[10,10])
```
