```python
import cv2
import matplotlib.pyplot as plt
import os

image_path = os.path.join("../images", "AI-01.jpeg")
image = cv2.imread(image_path)
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
# function to display the image
def show_image(title,img,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
show_image("Original image (BGR)",image)
negative_image = 255 - image
show_image("Negative image",negative_image)
```

```python
print("comparing pixel values")
print(image[10,10])
print(negative_image[10,10])
```

```python
#contrast stretching
image_path1 = os.path.join("../images", "low-contrast-image.jpg")
low_contrast_image = cv2.imread(image_path1)
low_contrast_image = cv2.cvtColor(low_contrast_image,cv2.IMREAD_GRAYSCALE)
```

```python
import numpy as np
min_pixel = np.min(low_contrast_image)
max_pixel = np.max(low_contrast_image)
high_contrast_image= ((low_contrast_image- min_pixel) / (max_pixel - min_pixel) * 255).astype(np.uint8)
```

```python
show_image("Low contrast",low_contrast_image)
show_image("High transformation",high_contrast_image)
```
