```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
image_path = os.path.join("../images", "salah.jpg")
image = cv2.imread(image_path)

# function to display the image
def show_image(title,img,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

show_image("Original image (BGR)",image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
show_image('Original Image(rgb)', image_rgb)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
adaptive_threshold = cv2.adaptiveThreshold(
    gray_image, 
    255,  # Maximum value to be assigned to pixels exceeding the threshold
    cv2.ADAPTIVE_THRESH_MEAN_C,  # Adaptive thresholding method
    cv2.THRESH_BINARY,  # Threshold type
    11,  # Block size (size of the neighborhood area)
    2  # Constant subtracted from the mean (fine-tuning parameter)
)
show_image('Binary Threshold Image', adaptive_threshold,cmap="gray")
```
