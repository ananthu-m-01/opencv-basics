```python
import cv2
import os
import matplotlib.pyplot as plt

image_path = os.path.join("../images/messi.webp")
image = cv2.imread(image_path)

# splitting into three channel
b_channel,g_channel,r_channel = cv2.split(image)

# function to display the image
def show_image(title,image,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(image,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
# merging r,g,b channel
merged_image = cv2.merge((r_channel,g_channel,b_channel))
show_image("Merged image(RGB)",merged_image)
```

```python
# merged BGR image
merged_BGRimage = cv2.merge((b_channel,g_channel,r_channel))
show_image("merged image(BGR)",merged_BGRimage)
```

```python
# merging GBR
merged_GBRimage = cv2.merge((g_channel,b_channel,r_channel))
show_image("merged image(GBR)",merged_GBRimage)
```
