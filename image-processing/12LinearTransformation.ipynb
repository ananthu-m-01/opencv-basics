```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np

# Define the image path
image_path = os.path.join("../images", "salah.jpg")

# Load the image in BGR format (default in OpenCV)
image_bgr = cv2.imread(image_path)

# Convert BGR to RGB for correct color visualization in Matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Load the grayscale image
gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Function to display the image
def show_image(title, img, cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

# Display the image in RGB format
show_image("Original Image (RGB)", image_rgb)
```

```python
alpha = 1.5  # Contrast control (1.0-3.0)
beta = 50  # Brightness control (0-100)
# Apply linear transformation (contrast and brightness adjustment)
linear_transformed = cv2.convertScaleAbs(gray_image, alpha=alpha, beta=beta)
show_image("Original image",image_rgb,cmap="gray")
show_image("Linear Transformed",linear_transformed,cmap="gray")
```
