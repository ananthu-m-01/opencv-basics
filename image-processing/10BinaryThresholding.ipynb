```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
image_path = os.path.join("../images", "salah.jpg")
image = cv2.imread(image_path)

# function to display the image
def show_image(title,img,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

show_image("Original image (BGR)",image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
show_image('Original Image(rgb)', image_rgb)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary_threshold = cv2.threshold(gray_image, 127, 255,cv2.THRESH_BINARY)
show_image('Binary Threshold Image', binary_threshold,cmap="gray")
```
