```python
import cv2
import matplotlib.pyplot as plt
import os

image_path = os.path.join("../images", "salah.jpg")
image = cv2.imread(image_path)

# function to display the image
def show_image(title,img,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

show_image("Original image (BGR)",image)
```

```python
g_channel, r_channel = image[:, :, 1], image[:, :, 2]
grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
show_image("Grayscale", grayscale, cmap="gray")
_,black_and_white = cv2.threshold(grayscale, 128, 255, cv2.THRESH_BINARY)
show_image("Black and White", black_and_white, cmap="gray")
custom_merge = cv2.merge((black_and_white, g_channel, r_channel))
show_image("Custom Merge (Black & White + G + R)", custom_merge)
```
