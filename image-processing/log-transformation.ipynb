```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

image_path = os.path.join("../images", "x-ray.webp")

# Load X-ray image in grayscale
xray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if image is loaded successfully
if xray_image is None:
    raise FileNotFoundError(f"Image not found at {image_path}")

# Function for displaying images
def show_image(title, img, cmap="gray"):
    plt.figure(figsize=(6, 6))
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

# Ensure image is not empty
max_pixel = np.max(xray_image)
if max_pixel == 0:
    raise ValueError("Image is completely black, log transformation is not possible.")

# Apply log transformation
c = 255 / np.log(1 + max_pixel)  # Compute scaling factor
log_transformed = c * np.log(1 + xray_image)  # Apply transformation

# Normalize result to 8-bit grayscale
log_transformed = np.array(log_transformed, dtype=np.uint8)

# Display images
show_image("Original X-ray", xray_image)
show_image("Log Transformed X-ray", log_transformed)
```
