```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np

# Define the image path
image_path = os.path.join("../images", "salah.jpg")

# Load the image in BGR format (default in OpenCV)
image_bgr = cv2.imread(image_path)

# Convert BGR to RGB for correct color visualization in Matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Load the grayscale image
gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Function to display the image
def show_image(title, img, cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

# Display the image in RGB format
show_image("Original Image (RGB)", image_rgb)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load grayscale image


# Apply histogram equalization
equalized_image = cv2.equalizeHist(gray_image)

# Create a figure with subplots
fig, ax = plt.subplots(2, 2, figsize=(12, 6))

# Display original grayscale image
ax[0, 0].imshow(gray_image, cmap='gray')
ax[0, 0].set_title("Original Image")
ax[0, 0].axis("off")

# Histogram of the original image
ax[0, 1].hist(gray_image.ravel(), bins=256, range=[0, 256], color='blue', alpha=0.7)
ax[0, 1].set_title("Histogram of Original Image")

# Display equalized image
ax[1, 0].imshow(equalized_image, cmap='gray')
ax[1, 0].set_title("Equalized Image")
ax[1, 0].axis("off")

# Histogram of the equalized image
ax[1, 1].hist(equalized_image.ravel(), bins=256, range=[0, 256], color='red', alpha=0.7)
ax[1, 1].set_title("Histogram of Equalized Image")

# Adjust layout and show the figure
plt.tight_layout()
plt.show()
```
