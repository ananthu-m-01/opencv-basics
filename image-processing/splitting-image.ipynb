```python
import cv2
import matplotlib.pyplot as plt
import os

image_path = os.path.join("../images", "messi.webp")
image = cv2.imread(image_path)

# splitting image into three channel
b_channel,g_channel,r_channel = cv2.split(image)
```

```python
# function to display image
def show_image(title,image,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(image,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
# b channel
show_image("B Channel ",b_channel,cmap="Blues")
```

```python
# red channel
show_image("Red Channel",r_channel,cmap="Reds")
```

```python
# green channel
show_image("Green channel",g_channel,"Greens")
```
