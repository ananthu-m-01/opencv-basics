```python
import cv2
import matplotlib.pyplot as plt
import os

image_path1 = os.path.join("../images", "AI-01.jpeg")
image1 = cv2.imread(image_path1)
image1 = cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)

image_path2 = os.path.join("../images", "AI-02.jpeg")
image2 = cv2.imread(image_path2)
image2 = cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
# function to display the image
def show_image(title,img,cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
def add_images(img1, img2):
    if img1.shape != img2.shape:
        img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
    added_image = cv2.add(img1, img2)

    return added_image
```

```python
added_result = add_images(image1, image2)

show_image("Original image 1",image1)
show_image("Original image 2 ",image2)
show_image("Added Image", added_result)
```

```python
print("pixel values")
print(image1[10,11])
print(image2[10,11])
# expected result is 26,99,114
print(added_result[10,11])
# The small difference in the Green channel (102 instead of 99) is likely due to internal optimizations in OpenCV.
```

```python
#XOR
image2 = cv2.resize(image2,(image1.shape[1],image1.shape[0]))
xor_image = cv2.bitwise_or(image1,image2)
show_image("XOR image",xor_image)
```

```python
print("comparison of pixel values")
print(image1[10,10])
print(image2[10,10])
print(xor_image[10,10])
```

```python
#Logical NOT
not_image1 = cv2.bitwise_not(image1)
show_image("Inverted images",not_image1)
```

```python
print("comparison of pixel values")
print(image1[10,10])
print(not_image1[10,10])
```
